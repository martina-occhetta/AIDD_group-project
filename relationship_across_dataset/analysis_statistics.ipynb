```python
import pandas as pd
```

# Counting under different filters

```python
df = pd.read_csv("AML_drugs_assay_target_disease_mapped_relationship_tidy.csv")
```

```python
df.head()
```

```python
df.sort_values(by=["parent_compound_specificity", "parent_compound_chembl_id"], ascending=[False, True], inplace=True)
```

```python
df.head()
```

```python
for idx, each_cid in enumerate(df['parent_compound_chembl_id'].unique()):
    df_temp = df[df['parent_compound_chembl_id'] == each_cid]
    print(idx, each_cid, df_temp['assay_chembl_id'].nunique(), df_temp['target_chembl_id'].nunique())
```

```python
print("In total", df['assay_chembl_id'].nunique(), df['target_chembl_id'].nunique())
```

```python
df_spec_accept = df[df['parent_compound_specificity'] != "wide-used"]
df_spec_accept.reset_index(drop=True, inplace=True)
```

```python
print("In total", df_spec_accept['assay_chembl_id'].nunique(), df_spec_accept['target_chembl_id'].nunique())
```

```python
df_uniprot = df_spec_accept[~df_spec_accept['OT_uniprot_id'].isna()]
df_uniprot.reset_index(drop=True, inplace=True)
```

```python
for idx, each_cid in enumerate(df_uniprot['parent_compound_chembl_id'].unique()):
    df_temp = df_uniprot[df_uniprot['parent_compound_chembl_id'] == each_cid]
    print(idx, each_cid, df_temp['assay_chembl_id'].nunique(), df_temp['target_chembl_id'].nunique())
```

```python
print("In total", df_uniprot['assay_chembl_id'].nunique(), df_uniprot['target_chembl_id'].nunique())
```

```python
df_AML_cell_lines = df_spec_accept[df_spec_accept['assay_cluster'] != "Not AML cell lines"]
df_AML_cell_lines.reset_index(drop=True, inplace=True)
```

```python
for idx, each_cid in enumerate(df_AML_cell_lines['parent_compound_chembl_id'].unique()):
    df_temp = df_AML_cell_lines[df_AML_cell_lines['parent_compound_chembl_id'] == each_cid]
    print(idx, each_cid, df_temp['assay_chembl_id'].nunique(), df_temp['target_chembl_id'].nunique(), df_temp['assay_cluster'].nunique())
```

```python
print("In total", df_AML_cell_lines['assay_chembl_id'].nunique(), df_AML_cell_lines['target_chembl_id'].nunique(), df_AML_cell_lines['assay_cluster'].nunique())
```

```python
df_uniprot_AML_cell_lines = df_AML_cell_lines[~df_AML_cell_lines['OT_uniprot_id'].isna()]
df_uniprot_AML_cell_lines.reset_index(drop=True, inplace=True)
```

```python
for idx, each_cid in enumerate(df_uniprot_AML_cell_lines['parent_compound_chembl_id'].unique()):
    df_temp = df_uniprot_AML_cell_lines[df_uniprot_AML_cell_lines['parent_compound_chembl_id'] == each_cid]
    print(idx, each_cid, df_temp['assay_chembl_id'].nunique(), df_temp['target_chembl_id'].nunique(), df_temp['assay_cluster'].nunique())
```

```python
print("In total", df_uniprot_AML_cell_lines['assay_chembl_id'].nunique(), df_uniprot_AML_cell_lines['target_chembl_id'].nunique(), df_uniprot_AML_cell_lines['assay_cluster'].nunique())
```

```python
'Not AML cell lines' in df_uniprot_AML_cell_lines['assay_cluster'].unique()
```

```python
for idx, each_cid in enumerate(df_spec_accept['parent_compound_chembl_id'].unique()):
    df_temp = df_spec_accept[df_spec_accept['parent_compound_chembl_id'] == each_cid]
    for jdx, each_assay_import in enumerate(df_temp['assay_importance'].unique()):
        df_temp_import = df_temp[df_temp['assay_importance'] == each_assay_import]
        num_assay = df_temp_import['assay_chembl_id'].nunique()
        num_target = df_temp_import['target_chembl_id'].nunique()
        num_cluster = df_temp_import['assay_cluster'].nunique()
        if "Not AML cell lines" in df_temp_import['assay_cluster'].unique():
            num_cluster -= 1
        print(idx, each_cid, each_assay_import, f"{num_assay} ({num_target}) \\newline ({num_cluster} clusters)")
```

```python
for important in df_spec_accept['assay_importance'].unique():
    df_spec_accept_import = df_spec_accept[df_spec_accept['assay_importance'] == important]
    
    num_assay = df_spec_accept_import['assay_chembl_id'].nunique()
    num_target = df_spec_accept_import['target_chembl_id'].nunique()
    num_cluster = df_spec_accept_import['assay_cluster'].nunique()
    if "Not AML cell lines" in df_spec_accept_import['assay_cluster'].unique():
        num_cluster -= 1

    print("In total", important, f"{num_assay} ({num_target}) \\newline ({num_cluster} clusters)")
```
